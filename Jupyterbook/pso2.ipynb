```python
# Inisialisasi partikel
import json

f = open('particle.json')

particle = json.load(f)

# print(particle)
f.close()

swarm = {}
for d in particle: #inisiasi partikel
    swarm[d] = {} # init dictonary

    swarm[d]['posisi'] = particle[d] # init posisi

    # init velocity
    swarm[d]['velocity'] = []  
    for p in range(len(swarm[d]['posisi'])):
        swarm[d]['velocity'].append(0)
    # swarm[d]['gbest'] = False
    swarm[d]['prev_fitness'] = 0
    swarm[d]['prev_posisi'] = []
    
    
# print(swarm)
```

```python
import random

def fitness(posisi):
    total_fitness = 0
    for i in posisi:
        iscore = info['kalimat ' + str(i)] # Ambil Iscore
        sum = 0
        for d in posisi[1:3]: # Ambil cosine_similarity
            if d == i:
                cos = 1 # posisi sama
            else:
                cos = cosine_similarity['kalimat ' +str(i)]['kalimat ' +str(d)]
            sum += iscore * cos
        total_fitness += sum
    return total_fitness


def find_iscore(pos):
    return info['kalimat '+ str(pos)]

def velocity(v0, xbest, gbest, xi, p):
    r1 = random.random()
    r2 = random.random()
    c1 = 2
    c2 = 2
    v = []
    for x in range(len(xi)):
        # Velocity
        # V = w*v0 + (c1*r1*(pBest - xi)) + (c2*r2*(gBest - xi))
        # V = velocity
        # w = inertia [0,1]
        # c1, c2 = cognitive [0,2]
        # r1, r2 = random [0,1]
        # xi = current position
        # pBest, gBest
        vt = 0.1*v0[x] + (c1*r1*(find_iscore(xbest[x]) - find_iscore(xi[x]))) + (c2*r2*(find_iscore(gbest[x]) - find_iscore(xi[x])))
        v.append(vt)
    swarm[p]['velocity'] = v
    # print(v)

def find_iscore_key(iscore):
    for i in info:
        if info[i] == iscore:
            return int(i[8:])

def update_posisi(pos, velo,p):
    atas = max(info, key=lambda x:info[x])
    bawah = min(info, key=lambda x:info[x])
    
    x_new = []
    for x in range(len(pos)):
        # print(find_iscore(pos[x]))
        # print(velo[x])
        xt = find_iscore(pos[x]) + velo[x]
        if xt > info[atas]:
            xt = info[atas]
            # i = 14
            # while xt == x_new[x-1] or xt == x_new[0]:
            #     xt = sort_info2[i-1]
        if xt < info[bawah]:
            xt = info[bawah]
            # i = 0
            # while xt == x_new[x-1]:
            #     xt = sort_info2[i+1]
        x_new.append(xt)
    # print(x_new)
    # s = 0
    # for x in x_new:
    #     s += x

    # avg = s/len(x_new)
    # new = []
    
    # for x in x_new:
    #     new.append(x/avg)


    new_pos = []
    for x in range(len(x_new)):
        temp = []
        for i in info:
            calc = info[i] - x_new[x]
            # if calc >= 0:
            temp.append((int(i[8:]), abs(calc))) # Absolut
        # s = min(temp, key=lambda x:x[1])
        # new_pos.append(s[0])
        # a_list = list(range(1, len(info))) # Simpan list iscore
        s = sorted(temp, key=lambda x:x[1])
        # print(s)
        # new_pos.append(s[0][0])
        # Kemungkinan kepakai
        for m in s:
            if m[0] not in new_pos:
                new_pos.append(m[0])
                break

            # elif m[0] in new_pos and m[1] >= 0:
            #     minus = []
            #     for n in s:
            #         if n[1] < m[1]:
            #             minus.append(n)
            #     close_to_zero = sorted(minus)
            #     get_zero = close_to_zero[-1:]

            #     new_pos.append(get_zero[0])

        
            
                

                
                    
        # print(s)
        # print(temp)
        # for t in temp:
        #     if t[1] == s[1]:
        #         new_pos.append(t[0])
        # for t in s:
        #     if t[1] == key2:
        #         if not new_pos:
        #             new_pos.append(t[0])
        #         elif key2 in new_pos:
        #             new_pos.append(s[1][0])
                        


        # for t in temp:
        #     if min(temp)[1] == t[1]:
        #         print(t)
    # print(new_pos)
    swarm[p]['posisi'] = new_pos
    print(p + ':' + str(new_pos))
    return swarm[p]['posisi']
```

```python

gbest_pos = []
gbest = 0
final = []
for it in range(1000):
    total_fitness = []
    for p in swarm:
        posisi = swarm[p]['posisi']
        fitness_partikel = fitness(posisi)
        total_fitness.append(fitness_partikel)

        swarm[p]['fitness'] = fitness_partikel


    # Gbest
    gbest = max(total_fitness)
    for p in swarm:
        if gbest == swarm[p]['fitness']:
            # swarm[p]['gbest'] = True
            gbest_pos = swarm[p]['posisi']
        else:
            # swarm[p]['gbest'] = False
            pass

    print("GBest iterasi " +str(it) + ": "+str(gbest))
    print("GBest Solusi : " + str(gbest_pos))
    
    if not final:
        final = gbest_pos
    elif final == gbest_pos:
        length_swarm = len(swarm)
        count = 0
        for p in swarm:
            if swarm[p]['posisi'] == gbest_pos:
                count += 1
        if count == length_swarm:
            break
    else:
        final = gbest_pos
        

    


    #Pbest
    for p in swarm:
        if swarm[p]['prev_fitness'] < swarm[p]['fitness']:
            swarm[p]['prev_fitness'] = swarm[p]['fitness']
            swarm[p]['prev_posisi'] = swarm[p]['posisi']


    # Update velocity
    for p in swarm:
        v0 = swarm[p]['velocity']
        pb = swarm[p]['prev_posisi']
        gb = gbest_pos
        xi = swarm[p]['posisi']
        velocity(v0, pb, gb, xi, p)
    # print(swarm)

    # Update Posisi
    for p in swarm:
        pos = swarm[p]['posisi']
        velo = swarm[p]['velocity']
        update_posisi(pos, velo, p)

final = sorted(final)
print("best position : ", final)
print("\n")
for f in final:
    get_kalimat = dic["kalimat " +str(f)]['kalimat']
    print(get_kalimat)
```
